```python
def FindIntegrationLimits(vpar):
    signal = np.sign(vpar)
    index = np.where(np.diff(signal) != 0)[0]
    return index

BouncePoints_15_RK4 = FindIntegrationLimits(PA15_RK4["vpar"])
BouncePoints_30_RK4 = FindIntegrationLimits(PA30_RK4["vpar"])
BouncePoints_45_RK4 = FindIntegrationLimits(PA45_RK4["vpar"])
BouncePoints_60_RK4 = FindIntegrationLimits(PA60_RK4["vpar"])
BouncePoints_90_RK4 = FindIntegrationLimits(PA90_RK4["vpar"])

BouncePoints_15_Boris = FindIntegrationLimits(PA15_Boris["vpar"])
BouncePoints_30_Boris = FindIntegrationLimits(PA30_Boris["vpar"])
BouncePoints_45_Boris = FindIntegrationLimits(PA45_Boris["vpar"])
BouncePoints_60_Boris = FindIntegrationLimits(PA60_Boris["vpar"])
BouncePoints_90_Boris = FindIntegrationLimits(PA90_Boris["vpar"])
```

```python
def SecondInvariant(vpar, v, t, BouncePoints):
    int = []
    for i in range(len(BouncePoints) - 1):
        aux = scipy.integrate.trapezoid(vpar.iloc[BouncePoints[i]:BouncePoints[i+1]]**2, t.iloc[BouncePoints[i]:BouncePoints[i+1]] )/v.iloc[BouncePoints[i]:BouncePoints[i+1]]
        int = np.concatenate((int, aux))
    return int
```

```python
I15_RK4 = SecondInvariant(PA15_RK4["vpar"], PA15_RK4["v"], PA15_RK4["t"], BouncePoints_15_RK4)
I30_RK4 = SecondInvariant(PA30_RK4["vpar"], PA30_RK4["v"], PA30_RK4["t"], BouncePoints_30_RK4)
I45_RK4 = SecondInvariant(PA45_RK4["vpar"], PA45_RK4["v"], PA45_RK4["t"], BouncePoints_45_RK4)
I60_RK4 = SecondInvariant(PA60_RK4["vpar"], PA60_RK4["v"], PA60_RK4["t"], BouncePoints_60_RK4)
I90_RK4 = SecondInvariant(PA90_RK4["vpar"], PA90_RK4["v"], PA90_RK4["t"], BouncePoints_90_RK4)

I15_Boris = SecondInvariant(PA15_Boris["vpar"], PA15_Boris["v"], PA15_Boris["t"], BouncePoints_15_Boris)
I30_Boris = SecondInvariant(PA30_Boris["vpar"], PA30_Boris["v"], PA30_Boris["t"], BouncePoints_30_Boris)
I45_Boris = SecondInvariant(PA45_Boris["vpar"], PA45_Boris["v"], PA45_Boris["t"], BouncePoints_45_Boris)
I60_Boris = SecondInvariant(PA60_Boris["vpar"], PA60_Boris["v"], PA60_Boris["t"], BouncePoints_60_Boris)
I90_Boris = SecondInvariant(PA90_Boris["vpar"], PA90_Boris["v"], PA90_Boris["t"], BouncePoints_90_Boris)
```
