```python
import numpy as np
import matplotlib.pyplot as plt
from methods import Compute_B, Compute_Vpar_Vperp
from scipy.constants import c, m_e, m_p
import scipy
import pandas as pd
```

```python
# constants
Re = 6378137
B0 = 3.07e-5                        
B0_Re3 = B0 * (Re**3)               
c2 = c**2    
```

```python
species = 'proton'
if species == 'proton':
    m = m_p
elif species == 'electron':
    m = m_e
energy = 1.0e+07
pitch_angle = 45
dt_boris = 1.0e-04
dt_rk4 = 1.0e-04

RK4_file = f"Lorentz/{species}/pitch_angle/{species}_trajectory_{energy:.1e}_{pitch_angle}_dt{dt_rk4:.1e}_RK4.csv"
Boris_file = f"Lorentz/{species}/pitch_angle/{species}_trajectory_{energy:.1e}_{pitch_angle}_dt{dt_boris:.1e}_Boris.csv"

RK4 = pd.read_csv(RK4_file)
Boris = pd.read_csv(Boris_file)
```

```python
RK4["Bx"], RK4["By"], RK4["Bz"] = Compute_B(RK4["x"], RK4["y"], RK4["z"])
Boris["Bx"], Boris["By"], Boris["Bz"] = Compute_B(Boris["x"], Boris["y"], Boris["z"])
```

```python
RK4["vpar"], RK4["vperp"] = Compute_Vpar_Vperp(RK4["vx"], RK4["vy"], RK4["vz"], RK4["Bx"], RK4["By"], RK4["Bz"])
Boris["vpar"], Boris["vperp"] = Compute_Vpar_Vperp(Boris["vx"], Boris["vy"], Boris["vz"], Boris["Bx"], Boris["By"], Boris["Bz"])
```

```python
# 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(RK4["x"]/Re, RK4["y"]/Re, RK4["z"]/Re, label='RK4')
# ax.plot(Boris["x"]/Re, Boris["y"]/Re, Boris["z"]/Re, label='Boris')
ax.set_xlabel('X [Re]')
ax.set_ylabel('Y [Re]')
ax.set_zlabel('Z [Re]')
# add Earth
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
x = np.outer(np.cos(u), np.sin(v))
y = np.outer(np.sin(u), np.sin(v))
z = np.outer(np.ones(np.size(u)), np.cos(v))
ax.plot_surface(x, y, z, color='b')

ax.set_xlim([-5, 5])
ax.set_ylim([-5, 5])
ax.set_zlim([-5, 5])
ax.legend()
plt.show()
```

```python
# 1st Invariant mu
RK4["v"] = np.sqrt(RK4["vx"]**2 + RK4["vy"]**2 + RK4["vz"]**2)
RK4["B"] = np.sqrt(RK4["Bx"]**2 + RK4["By"]**2 + RK4["Bz"]**2)
RK4["gamma"] = 1/np.sqrt(1 - (RK4["v"]**2/c2))

RK4["mu"] = RK4["gamma"] * m * RK4["vperp"]**2 / (2 * RK4["B"])

Boris["v"] = np.sqrt(Boris["vx"]**2 + Boris["vy"]**2 + Boris["vz"]**2)
Boris["B"] = np.sqrt(Boris["Bx"]**2 + Boris["By"]**2 + Boris["Bz"]**2)
Boris["gamma"] = 1/np.sqrt(1 - (Boris["v"]**2/c2))

Boris["mu"] = Boris["gamma"] * m * Boris["vperp"]**2 / (2 * Boris["B"])
```

```python
# [:int(Boris["t"].size/2)]
fig, ax = plt.subplots(2, 1, sharex=True)
ax[0].plot(RK4["t"], RK4["mu"], label = "RK4")
ax[0].plot(Boris["t"], Boris["mu"], label = "Boris") 
ax[0].set_ylabel("mu [J/T]")
ax[1].plot(RK4["t"], RK4["z"]/Re, label = "RK4")
ax[1].plot(Boris["t"], Boris["z"]/Re, label = "Boris")
ax[1].set_ylabel("z [Re]")
ax[1].set_xlabel("t")
plt.legend()
plt.show()
```

```python
def FindIntegrationLimits(vpar):
    signal = np.sign(vpar)
    index = np.where(np.diff(signal) != 0)[0]
    return index

BouncePoints_RK4 = FindIntegrationLimits(RK4["vpar"])
BouncePoints_Boris = FindIntegrationLimits(Boris["vpar"])
```

```python
def SecondInvariant(vpar, v, t, BouncePoints):
    int = []
    for i in range(len(BouncePoints) - 1):
        aux = scipy.integrate.trapezoid(vpar.iloc[BouncePoints[i]:BouncePoints[i+1]]**2, t.iloc[BouncePoints[i]:BouncePoints[i+1]] )/v.iloc[BouncePoints[i]:BouncePoints[i+1]]
        int = np.concatenate((int, aux))
    return int
```

```python
RK4_I = SecondInvariant(RK4["vpar"], RK4["v"], RK4["t"], BouncePoints_RK4)
Boris_I = SecondInvariant(Boris["vpar"], Boris["v"], Boris["t"], BouncePoints_Boris)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(np.linspace(0, RK4_I.size, RK4_I.size)/10000, RK4_I/Re, "--", label = "RK4")
plt.plot(np.linspace(0, Boris_I.size, Boris_I.size)/10000, Boris_I/Re, "--", label = "Boris")
# plt.ylim([1.5, 1.6])
plt.xlabel("t [s]")
plt.ylabel("I [Re]")
plt.title("Second Invariant")
plt.legend()
plt.show()
```
